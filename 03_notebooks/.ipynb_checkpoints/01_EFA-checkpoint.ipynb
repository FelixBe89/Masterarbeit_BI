## EFA

Im Folgenden führe ich nach dem allg. Standardprozess eine _explorative factor analysis_ (EFA) durch. Dazu nutze ich die Python-Bibliothek "FactorAnalyzer". 

### Quellen

Die wesentlichen Quellen zu diesem Vorgehen und der Auswertung sind: 
- [DataCamp Tutorial für eine EFA in Pytohn](https://www.datacamp.com/tutorial/introduction-factor-analysis)
- Die Monographie von: Pascal Jordan. (2019). Faktorenanalyse (1. Auflage, 2019). Edition Rainer Hampp. https://www.wiso-net.de/document/EBOK,AEBO__9783957103437175


```python
# Import der Standardmodule für eine EFA mit Python 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from factor_analyzer import FactorAnalyzer
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
from factor_analyzer.factor_analyzer import calculate_kmo

# Warnungen unterdrücken 
import warnings 
warnings.filterwarnings("ignore")

df = pd.read_csv(r"C:\Users\felix\OneDrive\Desktop\masterarbeit\01_data\01_csv_data\00_pi_basics\FINALLY_ALL_pi_data.csv", sep=",", encoding="utf-8")

df_numeric = df[["h_index", "publikationen_gesamt", "drittmittel_gesamt", "s2_citations", "drittmittel_volumen"]]
```

```python
# Bartlett-Test. Hier wird grds. ein Wert von < 0,05 erwartet 
_, p_value = calculate_bartlett_sphericity(df_numeric)
print(f"Bartlett-Ergebnis für den P-Wert: {float(p_value):.40f}")

# KMO-Test: Hier wird grds. ein Wert von über 0,6 erwartet 
kmo_all, kmo_model = calculate_kmo(df_numeric)
print(f"KMO Wert liegt bei: {float(kmo_model):.2f}")
```

```python
# Initialisiere Analyzer
fa = FactorAnalyzer(rotation=None)
fa.fit(df_numeric)

# Eigenwerte extrahieren
ev, v = fa.get_eigenvalues()

# Scree Plot erstellen, um Ellenbogenkante zu finden
plt.scatter(range(1, df_numeric.shape[1]+1), ev)
plt.plot(range(1, df_numeric.shape[1]+1), ev)
plt.title("Scree Plot")
plt.xlabel("Faktor")
plt.ylabel("Eigenwert")
plt.grid()
plt.show()
```

```python
# Faktoren anhand des Graphen auswählen und Ladungen anzeigen lassen 

fa = FactorAnalyzer(n_factors=2, rotation="promax", method="ml")
fa.fit(df_numeric)

# Ladungen anzeigen
loadings_df = pd.DataFrame(fa.loadings_, index=df_numeric.columns)
print(loadings_df, "\n\n")
```

## Zwischenfazit der EFA

Letztlich zeigt sich, dass nach der EFA eine Faktormenge mit n=2 fast die gesamte Varianz der Features erklärt! Die fünf numerischen Features teilen sich, wie durchaus auch zu erwarten war, in zwei Bereiche, die mit den Publikationen und Drittmitteln korrespondieren.

## Principal Component Analysis (PCA)

Zur Kontrastierung der EFA, die zwar nach den allg. Gütekriterien zulässig ist, aber doch intendiert, latente Variablen zu finden, kann eine PCA ergänzend sinnvolle Einblicke in die Strukturen und Wechselseitigkeiten der numerischen Features liefern. 

Für diese Analyse wurden folgende Quellen herangezogen: 
- ...

```python
# Import der Bibliotheken und Einlesen der Daten
import numpy as np 
from sklearn.decomposition import PCA 

df_pca = pd.read_csv(r"C:\Users\felix\OneDrive\Desktop\masterarbeit\01_data\01_csv_data\00_pi_basics\FINALLY_ALL_pi_data.csv", sep=",", encoding="utf-8")

df_pca_numeric = df[["h_index", "publikationen_gesamt", "drittmittel_gesamt", "s2_citations", "drittmittel_volumen"]]
```

```python
df_pca_scaled = (df_pca_numeric - df_pca_numeric.mean()) / df_pca_numeric.std()

print(df_pca_scaled.head())
```

```python
df_pca_scaled_model = PCA(n_components=2).fit_transform(df_pca_scaled)
```

```python
df_pca = pd.DataFrame(df_pca_scaled_model, columns=["Faktor1", "Faktor2"])

df_pca
```

```python
df_pca_scaled_model_vari = PCA(n_components=0.9).fit_transform(df_pca_scaled)

print("Anzahl der Komponenten für 90% Varianzerhalt: {}.".format(df_pca_scaled_model_vari.n_components_))
```

```python
print(df_pca_scaled_model_vari.explain_variance_ratio_)
```
